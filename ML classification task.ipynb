```python
one_hot_encoding_columns = ['has_credit_in_default','contact','p_outcome','last_month_contacted']
one_hot_encoding_data = df[one_hot_encoding_columns]
one_hot_encoding_data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(categories = 'auto', sparse = False, handle_unknown = 'error')
encoder.fit(one_hot_encoding_data)
category_set = encoder.transform(one_hot_encoding_data)
category_set = pd.DataFrame(category_set)
category_set.columns = encoder.get_feature_names_out()
category_set
```

```python
val_for_the_model = df[['age',
                          'last_contact_duration_secs',
                          'campaign',
                          'p_days',
                          'previous',
                          'emp_var_rate',
                          'cons_price_idx',
                          'cons_conf_idx',
                          'euribor3m',
                          'is_term_deposit']]
```

```python
data = pd.concat([category_set, val_for_the_model], axis=1)
data
```

```python
X = data.loc[:,~data.columns.isin(['is_term_deposit'])]

y = data.loc[:,'is_term_deposit']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train.scaled = scaler.fit_transform(X_train)
X_test.scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Logistic Regression regularisation

logreg = LogisticRegression()
logreg.fit(X_train.scaled, y_train)

y_pred_log = logreg.predict(X_test.scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score
```

```python
print(f"Accuracy score for LogisticRegression implementation in test: {accuracy_score(y_test,y_pred_log)}")
print(f"Accuracy score for LogisticRegression implementation in train: {accuracy_score(y_train,logreg.predict(X_train.scaled))}")
print(f"Confusion Matrix for LogisticRegression implementation:\n {confusion_matrix(y_test,y_pred_log)}")
print(f"Classification Report for LogisticRegression implementation:\n {classification_report(y_test,y_pred_log)}")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model_lr = LogisticRegression()

regularization_values = ['l1','l2']
solver_values = ['saga','liblinear']

param_grid_lr = dict(penalty = regularization_values, solver = solver_values)
grid_lr = GridSearchCV(estimator=model_lr, param_grid=param_grid_lr, scoring=None)
grid_result_lr = grid_lr.fit(X_train.scaled, y_train)

print('The best param are:', grid_result_lr.best_params_, 'The best score are:', grid_result_lr.best_score_)
                
```

```python
classifier_logistics = LogisticRegression(solver='saga',
                                          penalty='l2',
                                          fit_intercept=True,
                                         ).fit(X_train.scaled, y_train)

y_pred_train_logistics = classifier_logistics.predict(X_train.scaled)
y_pred_test_logistics = classifier_logistics.predict(X_test.scaled)

print('Test accuracy:{}'.format(accuracy_score(y_test, y_pred_test_logistics.round())) )
print('Train accuracy:{}'.format(accuracy_score(y_train, y_pred_train_logistics.round())) )     

print(classification_report(y_test, y_pred_test_logistics))
                                                                                   
                         
```

```python
from sklearn import svm
```

```python
estimator_svc = svm.SVC()
```

```python
import numpy as np
```

```python
kernels = ['linear', 'poly', 'rbf']

Cs = np.linspace(0.1,3,7) #The C parameter controls the trade-off between maximizing the margin (large C) and
#minimizing the classification error (small C). A smaller C encourages a larger margin but may allow some misclassification,
#while a larger C reduces the margin to ensure that more training points are classified correctly.

degrees = [2,3]

param_grid_sv = dict(kernel=kernels, C=Cs, degree=degrees)

grid_sv = GridSearchCV(estimator=estimator_svc, param_grid=param_grid_sv, scoring=None)

grid_result_sv = grid_sv.fit(X_train.scaled, y_train)

print('Best parameters are:', grid_result_sv.best_params_,'Best score is:', grid_result_sv.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn1 = KNeighborsClassifier()

knn1.fit(X_train.scaled, y_train)
y_pred_knn1 = knn1.predict(X_test.scaled)

print(f"Accuracy score for KNN implementation is: {accuracy_score(y_test,y_pred_knn1)}")
print(f"Accuracy score for KNN implementation in train is: {accuracy_score(y_train, knn1.predict(X_train.scaled))}")
print(f"Confusion Matrix for SVM implementation is:\n {confusion_matrix(y_test,y_pred_knn1)}")
print(f"Classification Report for SVM implementation is:\n {classification_report(y_test,y_pred_knn1)}")
```

```python
model_kneighbors = KNeighborsClassifier()

neighbors_values = [3,4,5]

p_values = [1,2] #p=1 manhattan distance, p=2 euclidean distance

param_grid_kn = dict(n_neighbors=neighbors_values, p=p_values)

grid_kn = GridSearchCV(estimator=model_kneighbors, param_grid=param_grid_kn, scoring=None)

grid_result_kn = grid_kn.fit(X_train.scaled, y_train)

print('Best parameters are:', grid_result_kn.best_params_,'Best score is:', grid_result_kn.best_score_)
```

```python
classifier_kneighbors = KNeighborsClassifier(n_neighbors=5,
                                            p=1).fit(X_train.scaled, y_train)

y_pred_train_kneighbors = classifier_kneighbors.predict(X_train.scaled)
y_pred_test_kneighbors = classifier_kneighbors.predict(X_test.scaled)

print('Test accuracy:{}'.format(accuracy_score(y_test, y_pred_test_kneighbors.round())) )
print('Train accuracy:{}'.format(accuracy_score(y_train, y_pred_train_kneighbors.round())) )     

print(classification_report(y_test, y_pred_test_kneighbors))
```

```python
def calculate_metrics(y_true, y_pred):
    return [accuracy_score(y_true, y_pred), precision_score(y_true, y_pred), recall_score(y_true, y_pred), f1_score(y_true, y_pred)]
```

```python
metrics_neighbors = calculate_metrics(y_test, y_pred_test_kneighbors)
metrics_logistics = calculate_metrics(y_test, y_pred_test_logistics)
metrics_svm = calculate_metrics(y_test, y_pred_svm)

df = pd.DataFrame({
    'Logistic Regression': metrics_logistics,
    'KNN': metrics_neighbors,
    'SVM': metrics_svm
}, index=['Accuracy', 'Precision', 'Recall', 'F1-score'])

df
```
